```python
#from matplotlib import pyplot as plt
#Inside matplotlib library having pyplot method which  assist us to plot the data and  as plt means small/nick name to call easily.

import matplotlib.pyplot as plt
import numpy as np
```

```python
#taking two numeric a and b data to plot

a=np.array([1,2,3,4,5,6,7])
b=np.array([1,2,3,4,5,6,7])
```

```python
plt.title("Linear Line")
plt.xlabel("a")
plt.ylabel("b")
plt.plot(a,b)
plt.show()

#plt.plot()method is there inside it here provide a and b data means value of the number which need to be plot.
#plt.show() it will show/plot  the graph.
```

```python
#we adding more variable g and h two variable numpy array created with any point and passing with the plot function and you can see the output here.

g=np.array([3,6,44,12,67,45,55])
h=np.array([22,4,5,6,78,22,12])
plt.xlabel("g")
plt.ylabel("h")
plt.plot(g,h)
plt.show()
```

* ### Note- The point id here Matplotlib.pyplot is having the plot method which need the argument to be put and through it i can plot the data and plt.show method is  up here with it.

```python
#BY ADDING 3RD PARAMETER (,'--g')  TO PLOT THE DATA GRAPH IN DOTED ANG G FOR REFLECTION OF  GREEN COLOR LINE 

plt.plot(g,h,'--g')
plt.show()
```

```python
#Bar Method :-

x=[5,8,10]
y=[12,16,6]
plt.bar(x,y)   #here using bar method and giving x and y value to it
plt.title('Bar graph') #plt title where we providing the bar graph here
plt.ylabel("Y axis")   #labeling x and y to appearing to x and y axis 
plt.xlabel("X axis")
plt.show()              #plt.show is show the data
```

## Histogram 

* A histogram takes in a series of data and divides the data into a number of bins, It can plot the frequency data points in each bin (i.e. the interval of points).It is useful in understanding the count of data ranges.When to use , we  should use histogram when we need count of the variable in a plot.

```python
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a, bins = [0,20,40,60,80,100]) #hist means histogram present here and we give data (a) and bins means giving any  intervals(0,20,40,...) to bins.
plt.title("histogram")
plt.show

#histogram created with finding out the frequency of the data
```


# The Iris Dataset
This data sets consists of 3 different types of irises'
(Setosa, Versicolour, and Virginica) petal and sepal
length, stored in a 150x4 numpy.ndarray

The rows being the samples and the columns being:
Sepal Length, Sepal Width, Petal Length and Petal Width.

The below plot uses the first two features.
See here <https://en.wikipedia.org/wiki/Iris_flower_data_set> for more
information on this dataset.


```python
#loading data (iris.csv)
dsiris=pd.read_csv('iris.csv')
print(dsiris)
print(dsiris.head())
sns.set(style="whitegrid")
ax=sns.stripplot(x='class',y='sepal length',data=dsiris)#here x=class providing and inside the class Setosa, Versicolour, and Virginica category are present and y=sepel length column and data=dsiris=iris.csv here. 
plt.title('Graph')
plt.show()
```

```python
#visualiation impact are change but rest of things are same as above
ax=sns.swarmplot(x='class',y='sepal length',data=dsiris)
plt.title('Graph')
plt.show()
```

```python
#here we changing alignment (x='petal length',y='class')
ax=sns.swarmplot(x='petal length',y='class',data=dsiris)
```
